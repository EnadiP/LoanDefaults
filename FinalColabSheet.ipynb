The data audit is used to understand the data better and its shortcomings. In this phase of the project the data was split into multiple smaller datasets in order to prepare it for the machine learning models. The data was vizualized using box plots and bar graphs. A light correlation analysis was also performed in order to understand the relationship of each feature to one another. 

After examanining the data, the data was split into testing and training, then further split into categorical and continous. This step was to ensure that when the data is cleaned each of the values are addressed differently and efficently. 

Categorical data is imputed using mode and continuous is imputed using median. Further steps include binning categorical featres and scaling, normalzing and removing outliers from the continuous features. 

After these steps are complete, the data was merged into 2 data frames, training and testing. 

This portion of the project is the longest and most time consuming since it prepares the dataset for the models to operate correactly.
